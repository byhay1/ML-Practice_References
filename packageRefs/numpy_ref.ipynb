```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [10, 7]
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 400)
y = np.tanh(x)
fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
x = np.linspace(0, 2*np.pi, 400)
y1 = np.tanh(x)
y2 = np.cos(x**2)
fig, axes = plt.subplots(1,2, sharey=True)
axes[1].plot(x,y1)
axes[1].plot(x,-y1)
axes[0].plot(x,y2)
```
